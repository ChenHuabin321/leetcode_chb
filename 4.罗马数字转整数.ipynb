```python
class Solution(object):
    def romanToInt(self, s):
        """
        :type s: str
        :rtype: int
        """
        nums = {
            'IV': 4,
            'IX': 9,
            'XL': 40,
            'XC': 90,
            'CD': 400,
            'CM': 900,
            'I': 1,
            'V': 5,
            'X': 10,
            'L': 50,
            'C': 100,
            'D': 500,
            'M': 1000
        }
        i = 0
        s_lst = []
        len_s = len(s)
        while i < len_s:
            s1 = s[i:i+2] # 
            if s1 in nums:
                s_lst.append(nums[s1])
                i += 2      #  2
                continue   # 
            s1 = s[i]  # 
            s_lst.append(nums[s1])
            i += 1
        return sum(s_lst)
```
