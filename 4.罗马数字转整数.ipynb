解题思路：将'IV'，'IX'，'XL'，'XC'，'CD'，'CM'，'I'，'V'，'X'，'L'，'C'，'D'，'M'等存储Wie字典，字典中value为对应的数值。对输入字符串进行遍历，从序号0开始，先取s[0:2]两个字符在nums1中查找，找到则将对应的数值加入列表保存，序号直接加2，没找到则取s[0]在nums2中查找，然后将找到的值加入列表保存，重复这一过程直到序号指导s的末尾。

```python
class Solution(object):
    def romanToInt(self, s):
        """
        :type s: str
        :rtype: int
        """
        nums = {
            'IV': 4,
            'IX': 9,
            'XL': 40,
            'XC': 90,
            'CD': 400,
            'CM': 900,
            'I': 1,
            'V': 5,
            'X': 10,
            'L': 50,
            'C': 100,
            'D': 500,
            'M': 1000
        }
        i = 0
        s_lst = []
        len_s = len(s)
        while i < len_s:
            s1 = s[i:i+2] # 取连续两个字符进行查找
            if s1 in nums:
                s_lst.append(nums[s1])
                i += 2      # 如果找到， 索引直接加2
                continue   # 注意，直接下一次循环
            s1 = s[i]  # 在连续两个字符没有找到的情况下，取一个字符进行查找
            s_lst.append(nums[s1])
            i += 1
        return sum(s_lst)
```

```python
s="MCMXCIV"
Solution().romanToInt(s)
```
