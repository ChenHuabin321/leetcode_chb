```python
class Solution(object):
    def reverse(self, x):
        """
        :type x: int
        :rtype: int
        """
        flag = 1
        x=str(x)
        if x.startswith('-'):
            flag = -1
            x = x[:0:-1]
        else:
            x = x[::-1]
        x = flag * int(x)
        return x if x>=-2**31 and x <=2**31-1 else 0
        
```
