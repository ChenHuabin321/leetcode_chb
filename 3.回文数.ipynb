本地解题是限制不能讲整数转化为字符串。

整体思路：从末尾（个位数）开始依次求每个位上的数字，并加入list中，最后合并list为一个字符串，转为字符串为整型再对比其值。

关键在于怎么获取每个位上的数字：num取余10就可以获得个位上的数字，然后num减去个位数在除以10，num原本十位的数字就变成了各位上的数字，重复这一过程指导除以10后的num结果为0.

```python
class Solution(object):
    def isPalindrome(self, x):
        """
        :type x: int
        :rtype: bool
        """
        if x < 0:
            return False
        x2 = x
        nums = []
        while True:
            temp = x2 % 10
            nums.append(temp)
            x2 = (x2 - temp) // 10
            if x2 == 0:
                break
        num_str = ''.join([str(i) for i in nums])
        num = int(num_str)
        if num == x:
            return True
        else:
            return False
                
              
```

```python
Solution().isPalindrome(111)
```
