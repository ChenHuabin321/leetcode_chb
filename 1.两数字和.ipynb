解法一：

遍历列表，用target减去当前元素，然后在当前元素之后的所有元素中查找，一定要注意，是在当前元素之后的所有元素中查找。

```python
def find_sum_target(nums, target):
    for i, v1 in enumerate(nums):
        v2 = target - v1
        try:
            j = nums[i+1:].index(v2)
        except Exception:
            continue
        return i, j+i+1
```

解法2：

创建一个以nums值为key，index为值的字典中，然后遍历nums，在字典中查找target与当前值的差，如果找到且不予当前值的索引相等则返回。

```python
def find_sum_target(nums, target):
    v_dict = {v1: i for i , v1 in enumerate(nums) }
    for i , v1 in enumerate(nums):
        j = v_dict.get(target-v1)
        if j is not None and i != j:
            return i, j
```
